### Граф вычислений

$$
y = X^2 + 5X + 7
$$

1. Посчитать $x^2$
2. Посчитать $5x$
3. Cложить этот результат и добавить 7

```python
X = torch.tensor(2.0, requires_grad=True)
y = X**2 + 5*X + 7
print(X)
print(y)
print(y.grad_fn)
```

```python
x = torch.tensor(6.0, requires_grad=True)
a = x*3
b = a+45
c = b**3
y = c
print(a)
print(b)
print(c)
print(y)

print(y.grad_fn)
print(y.grad_fn.next_functions)
```
