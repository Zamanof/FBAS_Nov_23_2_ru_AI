```python
import torch
```

### Тензор
1. Скаляр - одно число - 0 мерный массив
2. Вектор - список чисел - 1 мерный массив
3. Матрица - 2 мерный массив
4. Изображение (RGB) - 3 мерный массив
5. Пачка изображений - 4 мерный массив

```python
scalar = torch.tensor(65)
print(f"Scalar {scalar}, Shape: {scalar.shape}")

vector = torch.tensor([6, 5, 4, 32])
print(f"Vector {vector}, Shape: {vector.shape}")

matrix = torch.tensor([[25, 665], [23,44], [35, 52]])
print(f"Matrix {matrix}, Shape: {matrix.shape}")

tensor3D = torch.randn(5, 5, 5)
print(f"Tensor3D {tensor3D}, Shape: {tensor3D.shape}")
```

```python
# tensor generate
print(torch.zeros(5, 5))
print(torch.ones(5, 5))
print(torch.rand(5, 5))
print(torch.arange(0, 20, 2))
```

```python
# math operations
a = torch.tensor([1,2,3,4])
b = torch.tensor([5,6,7,8])
print(a+b) # сложение
print(a*b) # покомпонентное умножение
print(torch.dot(a, b)) # Скалярное произведение
```

```python
A = torch.tensor([[1,2],[3,4]])
B = torch.tensor([[5,6],[7,8]])
print(A@B)
```
