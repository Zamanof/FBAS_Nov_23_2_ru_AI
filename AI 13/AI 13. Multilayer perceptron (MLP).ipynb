```python
import numpy as np
from sklearn.neural_network import MLPClassifier
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
```

```python
model = MLPClassifier(hidden_layer_sizes=(5,), max_iter=6000, activation="tanh")
model.fit(X, y)
y_pred = model.predict(X)
print(y_pred)
```
