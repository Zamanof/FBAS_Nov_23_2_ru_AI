```python
import torch
import torch.nn as nn
```

```python
linear = nn.Linear(in_features=3, out_features=2)
X = torch.tensor([1.0,2.0,4.0])
y = linear(X)
print(f"X input {X}")
print(f"y output {y}")
print (f"y.shape {y.shape}")

print(f"W: {linear.weight}")
print(f"b: {linear.bias}")
```
