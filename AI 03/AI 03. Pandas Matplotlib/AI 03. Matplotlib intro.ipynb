```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x  = np.arange(0, 10)
```

```python
y = x *2
```

```python
plt.plot(x, y)
plt.title("Matolotlib intro")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0, 8)
plt.ylim(0, 16)
# plt.savefig("Matolotlib intro.png")
plt.show()
```
