```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([52, 57, 62, 70, 74])
```

```python
print(x)
```

```python
model = LinearRegression()
model.fit(x, y) # обучение модели по нашим данным
```

```python
a = model.intercept_
b = model.coef_[0]
print(f"Уравнение: y = {a:.2f} + {b:.2f} * x")
```

```python
y_pred = model.predict(x)
y_pred
```

```python
plt.scatter(x,y, color='red', label='Данные')
plt.plot(x, y, color='blue', label= "Регрессия")
plt.xlabel('Часы подготовки')
plt.ylabel('Оценка')
plt.title("Простая линейная регрессия")
plt.legend()
plt.show()
```

```python
while True:
    try:
        x_new = float(input("Введите часы подготовки"))
        if x_new == 0:
            print("Exit")
            break
        y_new = model.predict([[x_new]])[0]
        print(f"Prediction score = {y_new:.2f}")
        plt.scatter(x,y, color='red', label='Данные')
        plt.plot(x, y, color='blue', label= "Регрессия")
        plt.scatter(x_new,y_new, color='green', label='Ваш прогноз', zorder=5, s=100)
        plt.xlabel('Часы подготовки')
        plt.ylabel('Оценка')
        plt.title(f"Прогноз для {x_new} час")
        plt.legend()
        plt.show()
    except ValueError:
        print("Enter a number please")
```
